```python
category_df=pd.read_csv('List_csv\\product_category_name_translation.csv')
category_df.head()
```

```python
# Merge the Dataframe order_product and the Dataframe product_category_name_translation_df
product_Df = pd.merge(category_df,products_df,on='product_category_name', how='left')
product_Df.head(2)

# Del the product column from the product_catgeory_name
del product_Df['product_category_name']
# Del the product category 
del category_df

# Rename the Columns 
product_Df.rename(columns={'product_category_name_english': 'product_category'},inplace=True)
product_Df.head(2)
```

```python
df = pd.merge(order_df, customer_df, on='customer_id')
df = df.merge(items_df, on='order_id')
df = df.merge(payments_df, on='order_id')
df = df.merge(product_Df, on='product_id')
df = df.merge(sellers_df, on='seller_id')
df.head(10)

df.to_csv("merged_data.csv", index=False)
```

Customer By State

```python
import seaborn as sns
print(sns.__version__)
customer_by_state = df[['customer_unique_id', 'customer_state']].groupby('customer_state').count().reset_index()
customer_by_state = customer_by_state.sort_values(by=['customer_unique_id'])

sns.set_theme()
plt.figure(figsize=(15,10))
plt.bar(customer_by_state['customer_state'], customer_by_state['customer_unique_id'])
plt.show()
```

Number of order per year and per month

```python
# We have the new columns
df['order_purchase_year'] = pd.to_datetime(df['order_purchase_timestamp']).dt.year
df['order_purchase_month'] = pd.to_datetime(df['order_purchase_timestamp']).dt.month
df['order_purchase_day'] = pd.to_datetime(df['order_purchase_timestamp']).dt.day
df['order_purchase_hour'] = pd.to_datetime(df['order_purchase_timestamp']).dt.hour

orders = df[['order_id','order_purchase_year','order_purchase_month']]
orders= orders.groupby(['order_purchase_month','order_purchase_year'])['order_id'].count().reset_index()
orders = orders.sort_values(by=['order_purchase_year','order_purchase_month'])
orders['period'] = orders['order_purchase_month'].astype('str')+'/'+ orders['order_purchase_year'].astype(str)
orders.head(10)
```

```python
# Graph to Show the Data Representation

plt.figure(figsize=(10,8))
plt.bar(orders['period'],orders['order_id'])
plt.xticks(rotation = 75,fontsize =10,weight='bold')
plt.yticks(fontsize =10,weight='bold')
plt.show()
```

Top 10 Categories

```python
top_category = df[['product_category','order_item_id']]
top_category = top_category.groupby(['product_category']).sum().sort_values(by=['order_item_id'],ascending=False).reset_index()
top_category[:10]
```

```python
plt.figure(figsize=(15,10))
plt.bar(top_category['product_category'][:10], top_category['order_item_id'][:10])
plt.title('Number of products sold per category')
plt.xticks(rotation=75, fontsize=15, weight='bold')
plt.yticks(fontsize=15, weight='bold')
plt.show()
```

Count of order by status (Bar Chart)

```python
# Count the Number of order by status
order_by_status = df['order_status'].value_counts()

# Plot
plt.figure(figsize=(10,8))
colors = plt.cm.viridis(np.linspace(0, 1, len(order_by_status)))

# sns.barplot(x=order_by_status.index,y=order_by_status.values,palette ='viridis')
bars = plt.bar(order_by_status.index, order_by_status.values, color=colors)

# Adding labels on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,  # X-coordinate
             bar.get_height(),                 # Y-coordinate
             f'Qty: {int(bar.get_height())}',  # Label text
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.xticks(rotation =45)
plt.ylabel('Values of the Status')
plt.xlabel('Count of the Status')
plt.title('Number of the order by status')
plt.show()
```

Order over time in years (TIME SERIES)

```python
import locale
# Convert to datetime, handling errors
df['order_over_time'] = pd.to_datetime(df['order_purchase_timestamp'], errors='coerce')

# Drop rows where conversion failed
df = df.dropna(subset=['order_over_time'])

# Group by year
order_per_year = df['order_over_time'].dt.year.value_counts().sort_index()
# orders_per_day = df.groupby(df['order_over_time'].dt.year).size()

locale.setlocale(locale.LC_ALL,'en_IN.UTF-8')

# Plot
plt.figure(figsize=(12, 5))
plt.plot(order_per_year.index, order_per_year.values, marker='o', linestyle='-', color='b')

# Add labels on each point
for x, y in zip(order_per_year.index, order_per_year.values):
    plt.text(x, y+2000, f'Qty: {locale.format_string('%d',y,grouping=True)}'
            , ha='center', va='bottom', fontsize=10, fontweight='bold')

# Labels & Title
plt.xlabel("Year")
plt.ylabel("Number of Orders")
plt.title("Orders Trend Over Time")
plt.xticks(order_per_year.index)
plt.grid(axis='x')

plt.show()
```
